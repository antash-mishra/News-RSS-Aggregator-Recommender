```python
def recommend_similar_articles(news_id, num_recommendations=5):
    # Get the embedding for the given news article
    news_embedding = article_embeddings_df.loc[news_id].values.reshape(1, -1)
    
    # Calculate cosine similarity between the news article embedding and all other article embeddings
    similarity_scores = cosine_similarity(news_embedding, article_embeddings_df.values).flatten()
    
    # Create a dictionary mapping article IDs to their similarity scores
    article_similarity_scores = dict(zip(data_unique['clicked_news'], similarity_scores))
    
    # Find top N articles with highest similarity scores
    top_article_indices = heapq.nlargest(num_recommendations, range(len(similarity_scores)), key=similarity_scores.take)
    
    top_similarity_scores = [similarity_scores[idx] for idx in top_article_indices]

    
    # Return recommended articles and their similarity scores
    recommended_articles = data_unique.iloc[top_article_indices]
    recommended_articles['similarity_score'] = similarity_scores[top_article_indices]
    recommended_articles = recommended_articles.sort_values(by='similarity_score', ascending=False)

    return recommended_articles, top_similarity_scores, top_article_indices
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import string
import re
import spacy

# Download NLTK resources
nltk.download('punkt')
nltk.download('stopwords')


# Load spaCy English model
nlp = spacy.load("en_core_web_sm")

def preprocess_text(text):
    # Tokenization
    tokens = word_tokenize(text)

    # Lowercasing
    tokens = [token.lower() for token in tokens]

    # Remove punctuation
    tokens = [token for token in tokens if token not in string.punctuation]

    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [token for token in tokens if token not in stop_words]

    # Lemmatization using NLTK
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(token) for token in tokens]

    # Remove non-alphabetic characters
    tokens = [re.sub(r'[^a-zA-Z]', '', token) for token in tokens]

    # Remove empty tokens
    tokens = [token for token in tokens if token.strip() != '']

    # Join tokens back into text
    preprocessed_text = ' '.join(tokens)

    return preprocessed_text
```

```python
from annoy import AnnoyIndex
from scipy.sparse import csr_matrix
from tqdm import tqdm

with 
    def cosine_similarity_in_batches_approx(tfidf_matrix_sparse, batch_size=1000, n_neighbors=10):
        num_docs = tfidf_matrix_sparse.shape[0]
        num_batches = (num_docs + batch_size - 1) // batch_size  # Calculate number of batches

        # Initialize an empty cosine similarity matrix
        cosine_similarity_matrix = csr_matrix((num_docs, num_docs))

        # Initialize AnnoyIndex for approximate nearest neighbor search
        annoy_index = AnnoyIndex(tfidf_matrix_sparse.shape[1], 'angular')

        # Build AnnoyIndex with TF-IDF matrix
        for i, doc_vector in enumerate(tqdm(tfidf_matrix_sparse, desc="Building AnnoyIndex", unit="documents")):
            annoy_index.add_item(i, doc_vector.toarray()[0])

        annoy_index.build(n_neighbors)

        # Iterate over batches
        for i in tqdm(range(num_batches), desc="Computing Cosine Similarity", unit="batch"):
            start_idx = i * batch_size
            end_idx = min((i + 1) * batch_size, num_docs)
            batch_tfidf_matrix = tfidf_matrix_sparse[start_idx:end_idx]

            # Find approximate nearest neighbors for the batch
            neighbors_idx = []
            for doc_vector in batch_tfidf_matrix:
                nearest, _ = annoy_index.get_nns_by_vector(doc_vector.toarray()[0], n_neighbors, include_distances=False)
                neighbors_idx.append(nearest)

            # Compute cosine similarity between batch and neighbors
            batch_cosine_similarity = batch_tfidf_matrix.dot(tfidf_matrix_sparse[neighbors_idx].T)

            # Update the corresponding portion of the cosine similarity matrix
            cosine_similarity_matrix[start_idx:end_idx, neighbors_idx] = batch_cosine_similarity

        return cosine_similarity_matrix
```

```python
def cosine_similarity_batch(X_batch, Y):
    # Compute dot product between X_batch and Y
    dot_product = tf.matmul(X_batch, Y, transpose_b=True)
    # Compute norms of X_batch and Y
    norm_X = tf.norm(X_batch, axis=1, keepdims=True)
    norm_Y = tf.norm(Y, axis=1, keepdims=True)
    # Compute cosine similarity
    cosine_similarity_batch = dot_product / (norm_X * tf.transpose(norm_Y))
    return cosine_similarity_batch
```

```python
def compute_cosine_similarity_in_batches(X, Y, batch_size):
    num_batches = len(X) // batch_size + (1 if len(X) % batch_size != 0 else 0)
    cosine_similarity_batches = []
    for i in range(num_batches):
        start_idx = i * batch_size
        end_idx = min((i + 1) * batch_size, len(X))
        X_batch = X[start_idx:end_idx]
        cosine_similarity_batch = cosine_similarity_batch(X_batch, Y)
        cosine_similarity_batches.append(cosine_similarity_batch)
    return tf.concat(cosine_similarity_batches, axis=0)
```
